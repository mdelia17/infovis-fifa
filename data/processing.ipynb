```python
a.columns
```

```python
a = a.reset_index()
a = a.sort_values(by=['league_name', 'club_name'])
a_cols_float = ['punti_portiere','punti_attacco','punti_difesa','punti_tecnica','punti_velocita','punti_mentalita',
               'punti_potenza','overall']
a[cols_float] = a[cols_float].round(decimals=1)
a = a.reset_index()
a = a.drop(['index'],axis=1)
```
